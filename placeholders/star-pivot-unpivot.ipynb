```python
import pandas as pd
import seaborn as sns

star = pd.read_excel('datasets/star/star.xlsx')
star.head()
```

```python
star_pivot = pd.melt(frame=star, id_vars = ['schidkn', 'classk'], 
        value_vars=['tmathssk','treadssk'], var_name='test_type', value_name='score')
star_pivot.head()
```

```python
# Rename records in `test_type`

mapping = {'tmathssk':'math','treadssk':'reading'}

star_pivot['test_type'] = star_pivot['test_type'].map(mapping)
```

```python
star_pivot['test_type'].unique()
```

```python
star_pivot.pivot_table(index=['schidkn','classk'], 
                       columns='test_type', values='score').reset_index()
```
